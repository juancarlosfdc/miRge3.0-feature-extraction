```python
def scatter_hist(x, y, ax, ax_histx, ax_histy, line=True):
    ax.scatter(x, y)
    ax_histx.hist(x, bins = 20)
    ax_histy.hist(y, bins = 20, orientation='horizontal')
    if line:
        ax.plot([0, 1], [0,1], color='orange', transform = ax.transAxes)
        ax.text(.65, .45, r'line y = x', transform=ax.transAxes, size = 30, color='orange')
```
