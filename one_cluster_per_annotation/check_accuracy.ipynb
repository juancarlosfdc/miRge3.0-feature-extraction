# optionally: get a dataframe that was miRge training ready into shape for classification--basically, drop a couple of columns and generally preprocess it. It should have 74 columns. 

```python
novel_pileups_with_names = pd.read_csv('../hf_novel_pileups_features.ready_for_check_accuracy.csv')
```

```python
# df = pd.read_csv('../rnacentral_all_small_rnas_features_with_label.csv')
```

```python
featureListSelected = []
with open('../SVM_training/prec_only_no_pair_no_arm_features.txt', 'r') as inf:
    line = inf.readline()
    line = inf.readline()
    while line != '':
        print(line)
        featureListSelected.append((float(line.strip().split('\t')[1]), int(
            line.strip().split('\t')[2]), line.strip().split('\t')[0]))
        line = inf.readline()
print(featureListSelected)
data = novel_pileups_with_names.copy()
#data_x_raw = data.iloc[:,1:].values
#data_y_raw = data.iloc[:,0].values
#featureList_raw = data.columns.values.tolist()
#featureList_raw.insert(0, 'probability')
#featureList_raw.insert(0, 'predictedValue')
# Encode categorical features using one-hot-coding after deleting the 2nd and 3rd column in data
data.drop(data.columns[[1, 2]], axis=1, inplace=True)
data = pd.get_dummies(data)
data['clusterName'] = novel_pileups_with_names['clusterName']
#featureList = data.columns.values.tolist()[1:]

# subselect data with the selected features
totalfeatureList = data.columns.values.tolist()
subIndexList = [totalfeatureList.index(item[2]) for item in featureListSelected]
# Creat a variable for the feature data
data_x = data.iloc[:, subIndexList].values
```

```python
sc, clf, selectFeatureNameList = joblib.load('../SVM_training/rnacentral_all_small_rnas.pkl')
X_test_std = sc.transform(data_x)
preds = clf.predict(X_test_std)
novel_pileups_with_names['prediction_rnacentral_all_small_rnas'] = preds
```

```python
sc, clf, selectFeatureNameList = joblib.load('../SVM_training/miRGene_mRNA_train.pkl')
X_test_std = sc.transform(data_x)
preds = clf.predict(X_test_std)
novel_pileups_with_names['prediction_mirgene_exome'] = preds
```

```python
novel_pileups_with_names.to_csv('../novel_pileups_with_predictions.csv', index=None)
```
